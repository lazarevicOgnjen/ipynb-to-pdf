**Задатак 2.** а) Написати програмску секвенцу за рачунање факторијела $n!$ **природног броја** $n\in\mathbb{N}$ користећи `for` петљу. За вредност $n$ користити опцију корисничког уноса вредности `input`. Скицирати график функције на сегменту $[6,12].$

(4 поена)

```python
unosBroja = int(input("Unesi prirodan broj n: "))

rezultatFaktorijela = 1
for trenutniBroj in range(1, unosBroja + 1):
    rezultatFaktorijela *= trenutniBroj

print(f"Faktorijel broja {unosBroja} je: {rezultatFaktorijela}")

listaX = list(range(6, 13))
listaY = []

for broj in listaX:
    faktorijelTrenutni = 1
    for i in range(1, broj + 1):
        faktorijelTrenutni *= i
    listaY.append(faktorijelTrenutni)

plt.plot(listaX, listaY, marker='o')
plt.title("Grafik faktorijela na segmentu [6, 12]")
plt.xlabel("n")
plt.ylabel("n!")
plt.grid(True)
plt.show()
```

б) Написати програмску секвенцу за рачунање уопштеног факторијела $x!$ **позитивног броја** $x\in\mathbb{R}^+$ користећи `while` петљу. За вредност $x$ користити опцију корисничког уноса вредности `input`. Уопштени факторијел је функција 
$$x!=\begin{cases}x,&x\in(0,1],\\ x\cdot(x-1)!,&x>1.\end{cases}$$
Скицирати график функције на сегменту $[3.6,8.12].$

(6 поена)

```python
def izracunajUopsteniFaktorijel(ulazniBroj):
    if ulazniBroj <= 1:
        return ulazniBroj
    
    konacniRezultat = ulazniBroj
    while ulazniBroj > 1:
        ulazniBroj -= 1
        if ulazniBroj <= 1:
            konacniRezultat *= ulazniBroj
            break
        konacniRezultat *= ulazniBroj
    return konacniRezultat

korisnickiUnos = float(input("Unesite pozitivan broj x: "))
rezultatFunkcije = izracunajUopsteniFaktorijel(korisnickiUnos)
print(f"Uopšteni faktorijel broja {korisnickiUnos} je: {rezultatFunkcije}")

vrednostiX = np.arange(3.6, 8.12, 0.01)
vrednostiY = [izracunajUopsteniFaktorijel(tacka) for tacka in vrednostiX]

plt.plot(vrednostiX, vrednostiY)
plt.title("Uopšteni faktorijel x! na intervalu [3.6, 8.12]")
plt.xlabel("x")
plt.ylabel("x!")
plt.grid(True)
plt.show()
```

**Задатак 4.** Дат је низ случајно генерисаних бројева $v=\begin{bmatrix} x_0&x_1&\dots&x_n\end{bmatrix}$ (генерише се кодном ћелијом испод). Два елемента $x_i$ и $x_j$ у низу $v$ образују инверзију уколико је $i<j\ \wedge\ x_i>x_j.$ Написати ефикасан код којим се одређује број инверзија у низу $v.$

(10 поена)

```python
n=123
min, max=-10,10
v=rndm.uniform(min,max,n)
```

Као пример неефикасног решења дате су итерације. Можете да користите излаз ове функције за проверу тачности властитог кода. Искористите опцију мерења времена извршења да упоредите ефикасност алгоритма.

```python
start = timer()
inverzije=0
for i in range(n-1):
    for j in range(i+1,n):
        if v[i]>v[j]:
            inverzije+=1
print(inverzije)
end = timer()
print("Vreme izvrsenja prebrojavanja: ",end - start,"sec")            
```

```python
def brojInverzijaEfikasno(niz):
    def mergeSortSaBrojanjem(niz):
        if len(niz) <= 1:
            return niz, 0
        
        sredina = len(niz) // 2
        levaPolovina, brojInverzijaLevo = mergeSortSaBrojanjem(niz[:sredina])
        desnaPolovina, brojInverzijaDesno = mergeSortSaBrojanjem(niz[sredina:])
        
        spojeniNiz = []
        i = j = 0
        ukupnoInverzija = brojInverzijaLevo + brojInverzijaDesno
        
        while i < len(levaPolovina) and j < len(desnaPolovina):
            if levaPolovina[i] <= desnaPolovina[j]:
                spojeniNiz.append(levaPolovina[i])
                i += 1
            else:
                spojeniNiz.append(desnaPolovina[j])
                j += 1
                ukupnoInverzija += len(levaPolovina) - i
        
        
        spojeniNiz.extend(levaPolovina[i:])
        spojeniNiz.extend(desnaPolovina[j:])
        
        return spojeniNiz, ukupnoInverzija
    
    _, brojInverzija = mergeSortSaBrojanjem(niz.copy())
    return brojInverzija

pocetak = timer()
brojInverzijaEfikasno = brojInverzijaEfikasno(v)
kraj = timer()

print(f"Broj inverzija (efikasno): {brojInverzijaEfikasno}")
print(f"Vreme izvršenja efikasnog algoritma: {kraj - pocetak:.6f} sekundi")
```
